```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context('talk')
```

```python
!pwd
```

```python
df = pd.read_csv('../../zoobot/data/decals/classifications/streaming/aggregated.csv')
```

```python
df.columns.values
```

```python
df['smooth-or-featured_smooth_fraction'] = df['smooth-or-featured_smooth'] / df['smooth-or-featured_total-votes']
```

```python
fig, ax = plt.subplots()
sns.kdeplot(df['smooth-or-featured_smooth_fraction'], legend=False, ax=ax, bw=0.02)
# ax.hist(df['smooth-or-featured_smooth_fraction'], bins=15)
ax.set_xlabel(r'Smooth Vote Fraction $v$')
ax.set_ylabel('Density (Normalised)')
ax.set_xlim([0, 1])
fig.tight_layout()
fig.savefig('smooth_dist.png')
fig.savefig('smooth_dist.pdf')
```

```python
fig, ax = plt.subplots()
# sns.kdeplot(df['smooth-or-featured_total-votes'], legend=False, ax=ax, bw=2.)
ax.hist(df['smooth-or-featured_total-votes'], bins=200)
ax.set_xlabel(r'"Smooth?" Total Votes $v$')
ax.set_ylabel('Density (Normalised)')
ax.set_xlim([30, 50])
fig.tight_layout()
fig.savefig('decals_smooth_total_dist.png')
fig.savefig('decals_smooth_total_dist.pdf')
```

```python
fig, ax = plt.subplots()
sns.kdeplot(df['bar_total-votes'], legend=False, ax=ax)
# ax.hist(df['smooth-or-featured_smooth_fraction'], bins=15)
ax.set_xlabel(r'"Bar?" Total Votes $N_{bar}$')
ax.set_ylabel('Density (Normalised)')
ax.set_xlim([0, 40])
fig.tight_layout()
fig.savefig('decals_bar_total_dist.png')
fig.savefig('decals_bar_total_dist.pdf')
```
